```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython import display

%matplotlib notebook

# plt.figure(figsize=(10,5.625))
fig, ax = plt.subplots(figsize=(10,5.625))

angle = 12*180

# Set the range for the axis
x_limit = 8
y_limit = 4.5
y_shift = 2
ax.set_xlim(-x_limit, x_limit)
ax.set_ylim(-y_limit, y_limit)

# The x axis and y axis
# x axis
plt.plot([-x_limit,x_limit],[0,0],color = "w")
# y axis
plt.plot([0,0],[-y_limit,y_limit],color = "w")

# The function you want to draw
def X(t):
    y = np.sin(t)*(np.exp(np.cos(t)) - (2*np.cos(4*t)) - (np.sin(t/12))**5)
    return y

def Y(t):
    y = np.cos(t)*(np.exp(np.cos(t)) - (2*np.cos(4*t)) - (np.sin(t/12))**5)
    return y

t = np.linspace(0,angle*np.pi/180,angle + 1)
x_data = X(t)
y_data = Y(t)

def animation_frame(i):
    plt.plot(x_data[0:i],y_data[0:i],color = "#cc00cc",linewidth = 1)
    plt.xticks([])
    plt.yticks([])
    # Name = "k = "+str(k)
    # plt.title(Name)


# anim = FuncAnimation(fig, func=animation_frame, frames=range(0, 360+1, 1), interval=5)
anim = FuncAnimation(fig, func=animation_frame, frames=range(0, angle+1, 1), interval=10)
plt.show()

# writervideo = animation.FFMpegWriter(fps=50) #fps = Frame per second (This is the speed of the  animation) 
# file_name = "Butterfly.mp4"
# anim.save(file_name, writer=writervideo, dpi = 250) #dpi = dots per inch (This is the quality of the image)
```
